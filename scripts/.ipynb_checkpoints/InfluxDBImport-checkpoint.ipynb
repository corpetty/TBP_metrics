# TBP Specific functions and constants

````python
# Adding show differentiation
def get_show_title(title):
    if "BC" in title:
        return 'BlockChannel'
    elif 'On-Ramping' in title:
        return 'On-Ramping_with_Dee'
    elif 'Announcements' in title:
        return 'TBP_Announcements'
    elif 'NABP' in title:
        return 'Not_Another_Bitcoin_Podcast'
    elif 'Ether Review' in title:
        return "The_Ether_Review"
    elif 'Buy or Sell' in title:
        return "Buy_or_Sell,_What_the_Hell"
    elif 'An Ethereum' in title:
        return "An_Ethereum_Podcast"
    elif 'Humanist' in title:
        return 'Creating_a_Humanist_Blockchain_Future'
    elif "Hashing It Out" in title:
        return "Hashing_It_Out"
    elif "Crypto Until Infinity" in title:
        return "Crypto_Until_Infinity"
    else:
        return 'The_Bitcoin_Podcast'
````

# Create Helper Functions

````python
def dequote(s: str):
    """
    If a string has single or double quotes around it, remove them.
    Make sure the pair of quotes match.
    If a matching pair of quotes is not found, return the string unchanged.
    """
    if (s[0] == s[-1]) and s.startswith(("'", '"')):
        return s[1:-1]
    return s
````

````python
def create_json(measurement, title, show, source, update_time, count, release_date) -> dict:
    """
    Template for InfluxDB ingestion
    """
    return     {
        "measurement": measurement,
        "tags": {
            "title": title,
            "show": show,
            "source": source,
            "release_date": release_date,
        },
        "time": update_time,
        "fields": {
            "value": count
        }
    }
    
````

````python
def daily_json(measurement, source, date, count, update_time) -> dict:
    """
    Template for daily download InfluxDB ingestion
    """
    return {
        "measurement": measurement,
        "tags": {
            "update_time": update_time,
            "source": source,
        },
        "time": date,
        "fields": {
            "value": count
        }
    }
````

````python
def get_file_download_date(filename):
    """
    parses dates from format <year>-<month>-<date> from filename and returns datetime object.
    """   
    return dt.strptime(re.findall("([0-9]{4}\-[0-9]{2}\-[0-9]{2})", filename)[-1], '%Y-%m-%d')    
````

````python
def replace_ER_title(title):
    if '\u200a\u2014\u200a' in title:
        return title.replace('\u200a\u2014\u200a', ': ')
    elif '\u200a?"\u200a' in title:
        return title.replace('\u200a?"\u200a', ': ')
    else: 
        return title
    
def check_for_spotify(filename):
    if 'spotify' in filename:
        return "Spotify"
    else:
        return "LibSyn"
    
def format_release_date(date: str) -> str:
    return dt.strptime(re.findall("([0-9]{2}\/[0-9]{2}\/[0-9]{2})", date)[-1], '%m/%d/%y').isoformat(' ')
````

## New LibSyn Stats Exports

**NOTE:**  These stats overlap with legacy LibSyn stats but do not fully contain them, so they need to be checked against what currently exists, and create/update accordingly.  

Also, the `release_date` format has changed.  The format now passes through a unifying function to standardize across all formats.

````python
new_libsyn_files = []
dir_prefix = '../raw_data/'
for item in os.listdir(dir_prefix):
    if 'stats_TBP_episodic' in item:
        new_libsyn_files.append(dir_prefix + item)
````

````python
new_libsyn_files
````

````python
formatted_JSON = []
for filename in new_libsyn_files:
    print(filename)
    download_date = get_file_download_date(filename).isoformat(' ')
    df = pd.read_csv(filename)
    for index, row in df.iterrows():
        try:
            show = get_show_title(row.item_title)
            formatted_JSON.append(create_json(measurement='episode_downloads', 
                                              title=replace_ER_title(row.item_title), 
                                              show=show, 
                                              source=check_for_spotify(filename), 
                                              update_time=download_date, 
                                              count=row.downloads__total, 
                                              release_date=format_release_date(row.release_date)))
        except:
            show = get_show_title(row.Title)
            formatted_JSON.append(create_json(measurement='episode_downloads', 
                                              title=replace_ER_title(row.Title), 
                                              show=show, 
                                              source=check_for_spotify(filename), 
                                              update_time=download_date, 
                                              count=row.Downloads, 
                                              release_date=format_release_date(row.Released)))
````

````python
import pandas as pd
import requests

def get_track_downloads(show_name: str, client_id: str) -> pd.DataFrame:
    url = 'http://api.soundcloud.com/resolve?url=http://soundcloud.com/' + show_name + '/tracks&client_id=' + client_id
    tracks = requests.get(url=url)
    tracks = tracks.json()

    stats = []
    for track in tracks:
        url2 = "https://api-v2.soundcloud.com/tracks/" + str(track['id']) + "?client_id=" + client_id
        res = requests.get(url=url2).json()
        stats.append([res['title'], res['playback_count']])

    stats = pd.DataFrame(stats, columns=['title', 'playback_count'])
    stats['title'] = [dequote(title) for title in stats.title]
    stats = stats.set_index('title')
    return stats
````
